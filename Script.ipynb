```python
# Import OS Module

import os

# Set Directory where all files are present to rename
path=r"C:\Users\Admin\Desktop\Attacq\AFS_2017"
os.chdir(path)

# Print List of files in that folder and store the list into a variable
FolderList=os.listdir()

#For loop to check and rename using OS Builtin Functions
for i in FolderList:
    
    if i in FolderList:
        b="_Ext-own_"
        c=";Ext-own;"
        Imagename=i.replace(b,c)
        os.rename(i, Imagename)
        
    elif i in FolderList:
        d="_Int-own_"
        e=";Int-own;"
        Imagename=i.replace(d,e)
        os.rename(i, Imagename)
        
    elif i in FolderList:
        f="_Aerial-imp_"
        g=";Aerial-imp;"
        Imagename=i.replace(f,g)
        os.rename(i, Imagename)
        
    elif i in FolderList:
        h="_SV_"
        j=";SV;"
        Imagename=i.replace(h,j)
        os.rename(i, Imagename)
        
    else:
        print (i," not in the list")

        
```
